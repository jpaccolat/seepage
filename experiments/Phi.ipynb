Plot the dimensionless logarithmic sensitivity at psi=psi0 for the vGM parametrization. One can check that the it converges to $b$ at large suction whatever the shape parameter.

```python
import numpy as np
import matplotlib.pyplot as plt

# Internal imports
from rate import Phi
```

```python
# set plot style
%matplotlib widget
plt.style.use('plot_style.txt')
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

psi = np.logspace(-1.5, 1.5)
n = np.linspace(1.1, 6, 5)
for i, v in enumerate(n):
    y = Phi(psi, 1, v)
    b = (5 * v - 1) / 2
    ax.plot(psi, y / b, c=f'C{i}', label=f'$n={v:.2f}$')

ax.axhline(1, ls='-', c='k', lw=1)

ax.spines["top"].set_visible(True)
ax.spines["right"].set_visible(True)
ax.tick_params(top=True, right=True)
ax.set(xscale='log', yscale='log')
ax.minorticks_off()
ax.set_ylim(ymin=1e-5, ymax=1.5)

ax.set_xlabel('$\\psi / h_g$', fontsize=14)
ax.set_ylabel('$\\Phi / b$', fontsize=14)

plt.legend(loc='lower right', handlelength=0.8, columnspacing=0.8, fontsize=10)
plt.tight_layout()
```
